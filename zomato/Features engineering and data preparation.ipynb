From Exploration, Features such as country, city, Aggregate rating are most importance in predicting the If the resturant has online delivery or not.

```python
df.columns
```

```python
new_df = df[['City', 'Aggregate rating', 'Country', 'Has Online delivery']]
```

```python
new_df = new_df.rename(columns={'Aggregate rating':'Aggregate_rating', 'Has Online delivery': 'Online_delivery'})
```

```python
new_df.head()
```

```python
new_df.dtypes
```

```python
new_city = []

for i in df["City"]:
    new_city.append(i[0:3])
new_city
```

```python
len(new_city)
```

```python
new_city = pd.Series(new_city)

new_df.City = new_city


## removing rare values
temp_df = pd.Series(new_df['City'].value_counts())
temp_df[temp_df >= 0.1].index
```

```python
# let's create a dictionary to replace the rare labels with the
# string 'rare'

city_dict = {
    k: ('rare' if k not in temp_df[temp_df >= 0.2].index else k)
    for k in temp_df.index
}

city_dict
```

```python
city_grouped = new_df['City'].map(city_dict)
```

```python
dum = pd.get_dummies(city_grouped)
dum
```

```python
data_new = pd.concat([new_df, dum], axis=1)
```

```python
data_new
```

```python
data_new.drop("City", axis=1, inplace=True)
```

```python
dum = pd.get_dummies(data_new['Country'])
```

```python
data_new = pd.concat([data_new, dum], axis=1)
data_new.drop("Country", axis=1, inplace=True)
```
